## Bernoulli's Distribution

In a basket containing 100 fruits, there are 60 apples and 40 oranges. 

A fruit is picked randomly from the basket. 

1. **What is the probability that the fruit is apple?**
1. **What is the probability that it is an orange?**

````python
"""
Calculate:
p = probability that the fruit is an apple 
q = probability that the fruit is an orange
"""
````

````python
# Dados do problema
total_fruits = 100   # total de frutas
apples = 60          # quantidade de maçãs
oranges = 40         # quantidade de laranjas

# Cálculo das probabilidades
p = apples / total_fruits   # probabilidade de escolher uma maçã
q = oranges / total_fruits  # probabilidade de escolher uma laranja

# Arredondar para três casas decimais
p = round(p, 3)
q = round(q, 3)

# Exibir os resultados
print(f"Probability of picking an apple (p): {p}")
print(f"Probability of picking an orange (q): {q}")
````

## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python
# your solution here
import math

# parâmetros
n = 20   # total de frutas
k = 5    # número de maçãs
p = 0.6  # probabilidade de maçã
q = 0.4  # probabilidade de laranja

# fórmula binomial
prob = math.comb(n, k) * (p ** k) * (q ** (n - k))

# arredondar
prob = round(prob, 3)

print(f"Probability of getting 5 apples and 15 oranges: {prob}")

````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
# your code here
from scipy.stats import binom

# parâmetros
n = 20       # total de frutas
p = 0.6      # probabilidade de maçã

# CDF até k = 4 (menor que 5 maçãs)
prob_less_than_5 = binom.cdf(4, n, p)

# arredondar
prob_less_than_5 = round(prob_less_than_5, 3)

print(f"Probability that less than 5 apples are picked: {prob_less_than_5}")
````

<!-- **Plot the PDF (probability distributiom function) for the above binomial distribution.** -->

**Plot the Probability Mass Function (PMF) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
# your code here
# Please label the axes and give a title to the plot 

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# parâmetros
n = 20
p = 0.6

# valores possíveis de X (de 0 a n)
x = np.arange(0, n + 1)

# PMF da distribuição binomial
pmf_values = binom.pmf(x, n, p)

# plotagem
plt.figure(figsize=(8, 5))
plt.bar(x, pmf_values, color='skyblue', edgecolor='black')

# rótulos e título
plt.xlabel("Number of Apples Picked (k)")
plt.ylabel("Probability P(X = k)")
plt.title("Binomial Distribution PMF: n=20, p=0.6")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

```python
import math
math.factorial(n) # n is the number you want to find the factorial of
```

````python
# your code here 
import math

# parâmetros
lambd = 2.3   # média de gols por jogo
k = 5         # gols desejados

# fórmula da distribuição de Poisson
P = (math.exp(-lambd) * (lambd ** k)) / math.factorial(k)

# arredondar
P = round(P, 3)

print(f"Probability that exactly 5 goals are scored: {P}")
````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python
# your code here
# Please label the axes and give a title to the plot 
import math
import numpy as np
import matplotlib.pyplot as plt

# parâmetro
lambd = 2.3

# valores possíveis de gols (de 0 a 10)
x = np.arange(0, 11)

# função de probabilidade de Poisson (calculada manualmente)
poisson_pmf = [(math.exp(-lambd) * (lambd ** k)) / math.factorial(k) for k in x]

# plotagem
plt.figure(figsize=(8, 5))
plt.bar(x, poisson_pmf, color='lightgreen', edgecolor='black')

# rótulos e título
plt.xlabel("Number of Goals in a Match (k)")
plt.ylabel("Probability P(X = k)")
plt.title("Poisson Distribution (λ = 2.3): Goals per Match")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

````
